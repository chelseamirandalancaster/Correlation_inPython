```python
# import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

```

```python
# read in the data
df = pd.read_csv('/Users/ChelseaLancaster/Desktop/Projects/Movies/Python_Directory/movies.csv')
```

```python
# look at the data 
df.head()
```

```python
# Create a for loop to check for missing values 
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Look at data types
df.dtypes
```

```python
# change data type from float to integer in budget and gross
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
# create correct Year column 
df['yearcorrect'] = df['released'].astype(str).str[:4]
```

```python
# look at all of the data, not just 20 rows
pd.set_option('display.max_rows', None)

# order by gross revenue 
df = df.sort_values(by =['gross'], inplace = False, ascending = False)
```

```python
# drop any duplicates ; display all unique entries
df['company'].drop_duplicates().sort_values(ascending=False)

# drop duplicates across entire data set 
# df.drop_duplicates()
```

```python
# outcome predictions
# dependent variable : gross revenue
# predictors: budget (high correlation) ; company (high correlation)
```

```python
# build scatterplot to with budget v gross revenue
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Revenu')
plt.xlabel('Budget for Films')
plt.ylabel('Gross Revenue')
plt.show()
```

```python
df.head()
```

```python
# regression plot budget v gross using seaborn 

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "green"}, line_kws={"color":"purple"})
plt.xlabel("Budget")
plt.ylabel("Gross Revenue")
plt.title("Regression Plot")

# the regression plot shows there is a positive, moderate linear correlation
```

```python
# look at correlation matrix (only works on numerical fields, doesn't work on strings)

df.corr(method='pearson') # Pearson method ; default 

# note: high correlation between budget and gross 
```

```python
# visualize the correlation matrix 
correlation_matrix = df.corr(method='pearson')

# create heatmap
sns.heatmap(correlation_matrix, annot = True)
plt.title("Correlation Heatmap")
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")
plt.show()

# note: black is low correlation ; white-cream is high correlation
```

```python
# look at variable 'company' 
df.head()
```

```python
# give all string variables a unique identifier
df_numerized = df 

# for loop: if variable is an object, then turn it into a category
for col_name in df_numerized.columns: 
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
# visualize the correlation matrix 
correlation_matrix = df_numerized.corr(method='pearson')

# create heatmap
sns.heatmap(correlation_matrix, annot = True)
plt.title("Correlation Heatmap")
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")
plt.show()

# note: black is low correlation ; white-cream is high correlation
```

```python
# filtered correlation matrix 

df_numerized.corr()
```

```python
# unstack the matrix 
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
# further filter results by sorting pairs

sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
